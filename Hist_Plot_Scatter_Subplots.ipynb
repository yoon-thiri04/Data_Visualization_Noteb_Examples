```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#pdf of plot+histogram
mu,sigma=172,8
x=mu+sigma* np.random.randn(10000)
plt.hist(x,100,facecolor='green',density=True)
plt.xlabel('Heights')
plt.ylabel('Percentage of Students')
plt.title('Heights of the Students')
plt.grid(True)
plt.text(145,0.04,"u =172,sig=8")
plt.legend(['legend'])

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)

pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, 'k', linewidth=1)
plt.show()
```

```python
#plotting with two subplots
# Sample data for demonstration
x_data1 = [1, 2, 3, 4, 5]
y_data1 = [2, 4, 6, 8, 10]

x_data2 = [1, 2, 3, 4, 5]
y_data2 = [1, 3, 5, 7, 9]

#two subplots arranged in a 1x2 grid
fig, axs = plt.subplots(1, 2)

# Plot data for the first subplot with customization
axs[0].plot(x_data1, y_data1, label='Plot 1', marker='o', color='blue', linestyle='--')
axs[0].set_title('Subplot 1')
axs[0].set_xlabel('X-axis Label')
axs[0].set_ylabel('Y-axis Label')
axs[0].grid(True)
axs[0].legend(['Legend 1'])

# Add an annotation to the first subplot
axs[0].annotate('Annotation 1', xy=(2, 5), xytext=(3, 6),arrowprops=dict(facecolor='black', shrink=0.05))

# Plot data for the second subplot with customization
axs[1].plot(x_data2, y_data2, label='Plot 2', marker='s', color='green', linestyle='-.')
axs[1].set_title('Subplot 2')
axs[1].set_xlabel('X-axis Label')
axs[1].set_ylabel('Y-axis Label')
axs[1].grid(True)
axs[1].legend(['Legend 2'])

# Add an annotation to the second subplot
axs[1].annotate('Annotation 2', xy=(2, 4), xytext=(3, 6),arrowprops=dict(facecolor='black', shrink=0.05))

# Add an overall legend with custom text
fig.legend(['Overall Legend'], loc='upper right')

# Add a main title to the entire figure
plt.suptitle('Example Figure')

# Adjust layout to prevent clipping of the legend
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#plot figure with 4 subplots 
#creating 4 subplots >>> fig,axes=plt.subplots(2,2) Grid- x=2,y=2
#creating 2 subplots >>> fig,axes=plt.subplots(1,2) Grid- x=1,y=2 
fig,axes=plt.subplots(2,2)
axes[0,0].set_facecolor("green")
axes[0,1].set_facecolor('pink')
axes[1,0].set_facecolor('blue')
axes[1,1].set_facecolor('red')
```

```python
#sharex=True remains The margin label have only one for 2 axes
fig,axes=plt.subplots(2,sharex=True)
axes[0].set_facecolor('brown')
axes[1].set_facecolor('orange')
fig.subplots_adjust(hspace=.5)
```
